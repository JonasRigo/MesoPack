# Rate equation conductance

A very common master equation approach for quantum transport in QIMPs is the Wangsness-Block-Redfield (WBR) master equation. The concept of the master equation is based on the assumption that the eigenstates of the impurity $\{E_a,\ket{a}\}$ do not change when the bath gets connected. This assumption is justified for temperatures that are high compared to the hybridization strength and in particular, the approximation that is derived here limits the accessible temperatures to ${-\mathfrak{I}\Gamma(0)<T}$. If either the temperature gets very low or the hybridization very strong, the bath can quench impurity degrees of freedom as they become strongly entangled (as it is the case for the Kondo effect). We neglect such behaviour in the master equation approach and assume that impurity-bath states are product states at all times. At times $t \le 0$ we assume that the system is at equilibrium with $V_\alpha = 0$. At equilibrium the impurity can therefore be fully described by the Boltzmann distribution
$$
P_{t = 0}(a) = \frac{1}{\mathcal{Z}}e^{-\beta E_a} \;.
$$
When electrons are allowed to tunnel between the impurity and the bath ($V_\alpha > 0$), the impurity is no longer in equilibrium for leads with $\mu_s > \mu_d$ and a new \textit{non-equilibrium distribution} $P_{t > 0}(a)$ describes the distribution of eigenstates. This non-equilibrium distribution $P_{t > 0}(a)$ obeys the master equation
$$
\frac{d}{dt} P_t(a) = \sum_b \mathbb{\Gamma}_{a b} P(b) - P(a)\sum_{b}\mathbb{\Gamma}_{b a} \;,
$$
where $\mathbb{\Gamma}_{ab}$ are the Markov transition probabilities between impurity state $\ket{a}$ and $\ket{b}$, which depend on the hybridization between impurity and bath. The positive term in \ceqn{eq:trans-master-equation} describes the tunnelling in to state $\ket{a}$ and the negative term describes the tunnelling out of $\ket{a}$. If we express the occupation probability as a vector $[\mathbf{P}_t]_a = P_t(a)$, the master equation simplifies to 
$$
\frac{d}{dt} \mathbf{P}_t = [\mathbb{\Gamma}-\textrm{diag}(\mathbf{\Gamma})]\cdot\mathbf{P}_t\;,
$$
where $[\mathbf{\Gamma}]_a = \sum_b[\mathbb{\Gamma}]_{b a}$. 

There are several strategies to obtain the transition probabilities $\mathbb{\Gamma}$ \cite{timm2008tunneling}, but here we limit the discussion to the so-called "T-matrix approach". In the T-matrix approach we avail of an approximation of Fermi's golden rule to obtain the transition rates $\mathbb{\Gamma}$ between impurity-bath product states with different occupations. Hence, the tunnelling probability $[\mathbb{\Gamma}]_{b a}$ describes the rate with which the initial product state
$$
\ket{i^\alpha_{a_N}} \equiv \ket{N;a}\ket{i_\alpha} \;,
$$
transitions to a final product state with
$$
\ket{f^\alpha_{b_{N\pm1}}} \equiv\ket{N\pm1;b}\big(c^{(\dagger)}_{\alpha\sigma}\ket{i_\alpha}\big) \;,
$$
by moving an electron or a hole from the bath to the impurity. Here $\ket{i_\alpha}$ is the state of the of the source or drain lead, and $\ket{N;a}$ an impurity eigenstate with $N$ electrons. The transition rates between $\ket{i^\alpha_a}$ and $\ket{f^\alpha_b}$ are then given by the \tmat{}
$$
[\mathbb{\Gamma}^\alpha]_{b_{N\pm1}a_N } = 2\pi\sum_{if}| \bra{f^\alpha_{b_{N\pm1}}}\hat{T}\ket{i^\alpha_a}|^2 \times W_{i^a_\alpha}\delta(E_{N;a}+\epsilon^\alpha_i-E_{N\pm1;b}-\epsilon^\alpha_f) \;,
$$
where $\epsilon^\alpha_i$ are the single-particle energies of the lead $\alpha$ and $W_{i^a_\alpha}$ is the probability that the entire system is in the initial state $\ket{i^a_\alpha}$. The full transition probability is then given by given by the sum over the transition probabilities of the left and right lead $\mathbb{\Gamma} = \mathbb{\Gamma}^s + \mathbb{\Gamma}^d$. 

The T-matrix can be expanded as
$$
\hat{T} = \hat{H}_{\rm hyb} + \hat{H}_{\rm hyb}\frac{1}{\hat{H}_{\rm imp}-\hat{H}_{\rm leads}}\hat{H}_{\rm hyb} + \mathcal{O}(\hat{H}_{\rm hyb} ^3) \; ,
$$
when we restrict \ceqn{eq:master-rate} to the first order approximation of the T-matrix we get a convenient simple expression for the transition probabilities 
$$
[\mathbb{\Gamma}^\alpha]_{b_{N+1}a_N} = 2 \pi \sum_{i_\alpha} |\bra{N+1;b}\bra{i_\alpha} \hat{c}^{\dagger}_{\alpha\sigma} \hat{H}_{\rm hyb} \ket{N;a}\ket{i_\alpha}|^2\times W_{i^a_\alpha}\delta(E_{N;a}+\epsilon^\alpha_i-E_{N+1;b}-\epsilon^\alpha_f) \\
= 2\pi \rho_0 V^2_\alpha|\bra{N+1;b}\bar{\hat{d}}_{\alpha\sigma}\ket{N;a}|^2 \times f(E_{b_{N+1}} -E_{a_{N}} -\mu_\alpha)\;,
$$
where we have replaced the sum over bath states $\sum_{i_\alpha}$ with the integral over energies $\int d\epsilon \rho(\epsilon)$ and then used the wide flat bandwidth approximation. The above transition process describes an increase in impurity occupation, conversely the decreasing of impurity occupation is given by
$$
		[\mathbb{\Gamma}^\alpha]_{b_{N-1}a_N} = 2\pi \rho_0 V^2_\alpha|\bra{N-1;b}\bar{\hat{d}}_{\alpha\sigma}\ket{N;a}|^2 \times\left[1 -  f(E_{a_{N}} -E_{b_{N-1}} -\mu_\alpha)\right]\;.
$$
Thus, the entire information required to solve the master equation is encoded in the matrix elements $\bra{N\pm1;b}\bar{\hat{d}}_{\alpha\sigma}\ket{N;a}$ and Fermi-distribution of the bath. Note, that the baths are assumed to be at equilibrium at all times, this is justified by the vast size of the baths compared to the impurity. 

The final goal of the master equation is to compute the conductance across the impurity. Of particular interest is the conductance that sets it when $\mathbf{P}_t$ is stationary and thus, satisfies the steady state equation
$$
0 \stackrel{!}{=} [\mathbb{\Gamma}-\textrm{diag}(\mathbf{\Gamma})]\cdot\mathbf{P}_\infty \;,
$$
which can be solved by enforcing the normalization of the non-equilibrium distribution 
$$
	\sum_a [\mathbf{P}_\infty]_a = 1 \;.
$$
In the steady state the current across the nanostructure is given by
$$
I_C = -\sum_{N}\sum_{a,b} ([\mathbb{\Gamma}^s]_{b_{N+1}a_N}-[\mathbb{\Gamma}^s]_{b_{N-1}a_N})[\mathbf{P}_\infty]_{a_N} \;.
$$
which is identical but with opposite sign for $\alpha = d$ because of current conservation. Then, the conductance can be obtained from a derivative of $I_C$ with respect to $\mu_\alpha$.

## Coulomb diamond

The sequential tunnelling conductance is computed as a function of gate voltage and bias voltage.

```python
temp = 0.03

def fermi_parent(T):
    def fermi(x):
        return 1./(1+np.exp(x/T))
    return fermi

fd = fermi_parent(temp)
```

```python
U = .5
V = 0.137
t = 0.
N = 1
DOF = [4 for i in range(N)]
bDOF = [20 for i in range(1)]

p = {}
# [("hubbard-U",1),("extended-hubbard-U",1),("epsilon",1),("hopping",1),("magfield",1),("spin-spin",2)]
# the epsilon and U are to follow the order of the DOF
p["hubbard-U"] = np.array([U for i in range(N)]) #0.3
p["extended-hubbard-U"] = np.array([0. for i in range(N)]) #0.3
p["epsilon"] = np.array([-.5*U for i in range(N)]) #-.15
# the J and hopping are to follow the order of the bonds
p["spin-spin"] = np.array([[0.,0.] for i in range(N)])
p["hopping"] = np.array([t for i in range(N)])
# boundary conditions: are deactivated for now
# bcset = [1,[0,1]]
# creating the Hamiltonain class
H = InitHam(DOF,bonds = [])#,boson_DOF = bDOF)
# get the hamiltonian matrix
ham_neutral = H.getH(p)#,boson_site_coupling=[[0,0,0.1]],
            #boson_bond_coupling=[[0,0,1.0]],
            #boson_omega=[1e-4])
q = {}
q["epsilon"] = np.array([1. for i in range(N)])
n_all = H.getH(q)
#print(ham.GetMat(0,False))
```

```python
#####################
# initializing operators
#####################
# initializing the observables
SysObs = Operators(H)
# constructing the observables
Stot = SysObs.Stot()
n_up = SysObs.anihilation(0,"up")
cd_up = SysObs.creation(0,"up")
c_up = SysObs.anihilation(0,"up")
n_do = SysObs.anihilation(0,"down")
cd_do = SysObs.creation(0,"down")
c_do = SysObs.anihilation(0,"down")
# multiply cd and c
n_up.OperatorDot(cd_up)
n_do.OperatorDot(cd_do)
# one can create independent copies of operators
# note that one has to call the copy function
n = n_up.copy()
n.OperatorSum(n_do)
# conductance class is initialized
left_hyb = [V]
right_hyb = [V]
cccond = CalcCond(SysObs,left_hyb ,right_hyb)
```

```python
n_V = 120
V_set_1 = np.linspace(-1.5*U,1.5*U,n_V)
V_set_2 = np.linspace(-1.5*U,1.5*U,n_V)
#V_set_1 = V_set_2
Current = np.zeros((2,len(V_set_2),len(V_set_1)))
Current2 = np.zeros((2,len(V_set_2),len(V_set_1)))
precision = np.zeros((len(V_set_2),len(V_set_1)))
FinDiff = 1e-7
```

```python
for i1,Vgate in enumerate(V_set_2):
    # add the gate potential
    ham = ham_neutral.copy()
    eps_V = n_all.copy()
    eps_V.ScalarDot(Vgate)
    ham.OperatorSum(eps_V)
    # diagonalize 
    a, b = ham.Diagonalize(False)
    # temperature loop
    for j1,LeftVoltage_ in enumerate(V_set_1):
        # setting the fermi function
        fd = fermi_parent(temp)
        # impurity partition function
        Z0 = sum([np.sum(np.exp(-x[1]/temp)) for x in a])
        # compute occupation
        current = [0,0]
        for fd_ind,fd_sign in enumerate([-1,1]):
            # finite difference
            LeftVoltage = LeftVoltage_+fd_sign*FinDiff
            # Gamma matrix
            Gamma_Mat,Gamma_in,Gamma_out,P_init = cccond.GammaMat(a,b,LeftVoltage,temp)
            #trafo = np.dot(w,np.dot(v_limit,w.T))
            P_final,_ = cccond.rk4(Gamma_Mat, P_init, N=100)#np.dot(trafo,intt_prob_vec)
            P_final /= sum(P_final)
            # compute conductance 
            Current[fd_ind,i1,j1] = cccond.Current(Gamma_in,Gamma_out,P_final)
# compute differential conductance
Cond = (Current[0]-Current[1])/(2.*FinDiff)
```

```python
######################################
# plotting
######################################
fig = plt.figure()
fig, ax = plt.subplots(1,1, figsize=(4.3,3.4),dpi=150)

pcm = ax.pcolormesh(V_set_2,V_set_1, Cond.T, cmap='Blues',shading ='gouraud',vmin=0.05)
# Add a colorbar to the plot
cbar = plt.colorbar(pcm)
cbar.set_label('$G_C/G_0$',fontsize=labelfont)
cbar.ax.tick_params(labelsize=tickfont-2) 
ax.set_xlabel("$V_{\\rm gate}$",fontsize=labelfont)
ax.set_ylabel("$V_{\\rm bias}$",fontsize=labelfont)
ax.tick_params(axis='y', which='major', labelsize=tickfont)
ax.tick_params(axis='y', which='minor', labelsize=tickfont)
ax.tick_params(axis='x', which='major', labelsize=tickfont)
ax.tick_params(axis='x', which='minor', labelsize=tickfont)
```

## Temperature dependent conductance

```python
###########################
# defining the voltange set
###########################
n_V = 250
LeftVoltage_ = 0.
V_set_2 = np.linspace(-1.5*U,1.5*U,n_V)
temp_set = list(reversed(sorted([i*j for i in reversed(range(2,11)) for j in [0.1,0.01]])))
print(temp_set)                                 
Current = np.zeros((2,len(V_set_2)))
Current2 = np.zeros((2,len(V_set_2)))
Cond = np.zeros((len(temp_set),len(V_set_2)))
occupation = np.zeros((len(temp_set),len(V_set_2)))
accuracy = np.zeros((len(V_set_2)))
FinDiff = 1e-8
```

```python
for ind,Vgate in enumerate(V_set_2):
    # add the gate potential
    ham = ham_neutral.copy()
    eps_V = n_all.copy()
    eps_V.ScalarDot(Vgate)
    ham.OperatorSum(eps_V)
    # diagonalize 
    a, b = ham.Diagonalize(False)
    # temperature loop
    for tind,temp in enumerate(temp_set):
        # setting the fermi function
        fd = fermi_parent(temp)
        # impurity partition function
        Z0 = sum([np.sum(np.exp(-x[1]/temp)) for x in a])
        # compute thermal occupation expectation value
        occupation[tind,ind] = sum([np.sum(np.diag(n_all.BraKetDot(x,x)[1])*np.exp(-y[1]/temp)/Z0) for x,y in zip(b,a)])
        #continue
        for fd_ind,fd_sign in enumerate([-1,1]):
            # finite difference
            LeftVoltage = LeftVoltage_+fd_sign*FinDiff
            # Gamma matrix
            Gamma_Mat,Gamma_in,Gamma_out,P_init = cccond.GammaMat(a,b,LeftVoltage,temp)
            #trafo = np.dot(w,np.dot(v_limit,w.T))
            P_final,_ = cccond.rk4(Gamma_Mat, P_init, N=100)#np.dot(trafo,intt_prob_vec)
            P_final /= sum(P_final)
            # compute conductance 
            I = cccond.Current(Gamma_in,Gamma_out,P_final)
            Current[fd_ind,ind] = I
            # compute differential conductance
        Cond[tind,ind] = (Current[0,ind]-Current[1,ind])/(2.*FinDiff)
```

```python
######################################
# plotting
######################################

tickfont = 12
labelfont = 11

fig = plt.figure()
fig, bx = plt.subplots(1,2, figsize=(8,4),dpi=150)

# color map
cmap = plt.cm.cool
sm = plt.cm.ScalarMappable(cmap=cmap,norm=colors.LogNorm(vmin=temp_set[-1], vmax=temp_set[0]))
colorbar = plt.colorbar(sm)
colorbar.ax.set_title("$T$")

ax = bx[0]
ax.set_title("$\\rm conductance$",size=tickfont)
for i in range(0,len(temp_set)): ax.plot(V_set_2,\
                                         0.5*(Cond[len(temp_set)-i-1]+Cond2[len(temp_set)-i-1]),color=cmap(i / len(temp_set)))
ax.tick_params(axis='both', which='major', labelsize=tickfont)
ax.tick_params(axis='both', which='minor', labelsize=tickfont)
ax.set_ylabel("$G_C(0,T)/G_0$",fontsize=labelfont)
ax.set_xlabel("$V_{\\rm gate}$",fontsize=labelfont)
ax.tick_params(axis='y', which='major', labelsize=tickfont)
ax.tick_params(axis='y', which='minor', labelsize=tickfont)
ax.tick_params(axis='x', which='major', labelsize=tickfont)
ax.tick_params(axis='x', which='minor', labelsize=tickfont)

# Next panel
ax = bx[1]
ax.set_title("$\\rm occupation$",size=tickfont)
for i in range(0,len(temp_set)): ax.plot(V_set_2,occupation[len(temp_set)-i-1],color=cmap(i / len(temp_set)))
ax.tick_params(axis='both', which='major', labelsize=tickfont)
ax.tick_params(axis='both', which='minor', labelsize=tickfont)
ax.set_xlabel("$V_{\\rm gate}$",fontsize=labelfont)
ax.tick_params(axis='y', which='major', labelsize=tickfont)
ax.tick_params(axis='y', which='minor', labelsize=tickfont)
ax.tick_params(axis='x', which='major', labelsize=tickfont)
ax.tick_params(axis='x', which='minor', labelsize=tickfont)
ax.set_yticks([0,1,2])
```
